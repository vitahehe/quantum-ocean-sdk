```python
# This notebook demonstrates the Hamiltonian generators from the Hamiltonian
from Hamiltonian import (
    generate_hamiltonian,
    generate_hamiltonian_optimized,
    generate_hamiltonian_optimizedG,
    generate_hamiltonian_optimizedPAR,
)
```

```python
# build simple event and compare generators
N_MODULES = 3
LX = float("+inf")
LY = float("+inf")
Z_SPACING = 1.0

detector = toy.SimpleDetectorGeometry(
    module_id=list(range(N_MODULES)),
    lx=[LX] * N_MODULES,
    ly=[LY] * N_MODULES,
    z=[i + Z_SPACING for i in range(N_MODULES)],
)

generator = toy.SimpleGenerator(detector_geometry=detector, theta_max=np.pi/6)
event = generator.generate_event(2)

params = {"lambda":1.0, "alpha":0.5, "beta":0.1}

A1, b1, _ = generate_hamiltonian(event, params)
A2, b2, _ = generate_hamiltonian_optimized(event, params)
A3, b3, _ = generate_hamiltonian_optimizedPAR(event, params)

print("shapes", A1.shape, A2.shape, A3.shape)
```
